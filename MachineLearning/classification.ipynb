```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import sklearn as sk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("housing.csv")
data = data.dropna()
data.head()
```

```python
label_encoder = LabelEncoder()
data.insert(1, "Type 2 Classes", label_encoder.fit_transform(data["ocean_proximity"]), True)
data
```

```python
X = data[["median_house_value"]]
y = data["Type 2 Classes"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
np.array(X_test)
```

```python
np.array([[452600.0, 358500.0, 352100.0, 341300.0, 342200.0]]).shape
```

```python
y_pred = classifier.predict(X_test)
y_true = np.array(data["Type 2 Classes"])[len(data["Type 2 Classes"]) - len(y_pred):]
matches = []
number_of_trues = 0
number_of_falses = 0
for i, item in enumerate(y_pred):
    if item == y_true[i]:
        matches.append(True)
        number_of_trues += 1
    else:
        matches.append(False)
        number_of_falses += 1

matches, number_of_trues, number_of_falses, len(y_pred), len(y_true)
```

```python
print(f"Accuracy Score: {accuracy_score(y_true=y_true, y_pred=y_pred)}")
print(f"F1 Score: {f1_score(y_true=y_true, y_pred=y_pred, average="micro")}")
```
